```python
import pandas as pd
import numpy as np
```

# Read Data

```python
fn_train = 'aps_failure_training_set.csv'
fn_test = 'aps_failure_test_set.csv'

train = pd.read_csv(fn_train, skiprows=np.arange(20))
test = pd.read_csv(fn_test, skiprows=np.arange(20))

train_data = train.iloc[:, 1:]
train_label = train['class']

test_data = test.iloc[:, 1:]
test_label = test['class']
```

```python
train_data
```

```python
test
```

# Preprocessing

## imputation 

```python
from sklearn.preprocessing import Imputer
```

```python
def replace_(data, src, des):
    data_replaced = [[ des if c == src else c for c in r] for r in data]
    return pd.DataFrame(data_replaced)

train_replaced = replace_(train_data.values, 'na', np.nan)
test_replaced = replace_(test_data.values, 'na', np.nan)

imputer = Imputer(missing_values='NaN', strategy='mean')
imputer.fit(train_replaced)
train_imputed = pd.DataFrame(imputer.transform(train_replaced))
train_imputed.columns = train_data.columns

test_imputed = pd.DataFrame(imputer.transform(test_replaced))
test_imputed.columns = test_data.columns
```

# Model selection

## Ramdom Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
from sklearn.metrics import accuracy_score

import matplotlib.pyplot as plt
```

```python
rfc = RandomForestClassifier()
rfc.fit(train_imputed, train_label)
test_label_predicted = rfc.predict(test_imputed)

tn, fp, fn, tp = confusion_matrix(test_label, test_label_predicted).ravel()

tn, fp, fn, tp
```

### test error

```python
accuracy = accuracy_score(test_label, test_label_predicted)

'test effor: %.5f' % (1-accuracy)
```

```python
test_label_predicted_pro = rfc.predict_proba(test_imputed)

fpr, tpr, _ = roc_curve(test_label, test_label_predicted_pro[:, 1], pos_label='pos')

fpr, tpr

roc_auc = auc(fpr, tpr)

roc_auc
```

```python
plt.figure()
plt.plot(fpr, tpr, label='AUC %0.5f' % roc_auc)
plt.plot([0, 1], [0, 1])
plt.xlabel('False Positive rate')
plt.ylabel('True Positive rate')
plt.legend(loc='lower right')
plt.grid(True)

plt.show()
```

### out of bag error

```python
rfc = RandomForestClassifier(oob_score=True)
rfc.fit(train_imputed, train_label)

'out of bag error %.5f' % (1-rfc.oob_score)
```

# Deal imbanced data in RamdomForest model

```python
rfc = RandomForestClassifier(class_weight='balanced')
rfc.fit(train_imputed, train_label)
test_label_predicted = rfc.predict(test_imputed)

tn, fp, fn, tp = confusion_matrix(test_label, test_label_predicted).ravel()
tn, fp, fn, tp
```
