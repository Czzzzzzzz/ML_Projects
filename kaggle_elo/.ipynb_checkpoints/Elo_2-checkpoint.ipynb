Explore the correlation between number of transactions and purchase_aount

```python
gdf = hist_trans.groupby("card_id")
gdf = gdf['purchase_amount'].size().reset_index()
gdf.columns = ['card_id', 'number_transactions']

train_df = pd.merge(train, gdf, on="card_id", how="left")
test_df  = pd.merge(test, gdf, on="card_id", how="left")
```

'purchase_amount' in historical transaction

```python
gdf = hist_trans.groupby("card_id")

purchase_amount = gdf['purchase_amount'].agg(['sum', 'mean', 'std', 'min', 'max']).reset_index()
purchase_amount.columns = ['card_id', 'hist_pa_sum', 'hist_pa_mean', 'hist_pa_std', 'hist_pa_min', 'hist_pa_max']

train_df = pd.merge(train_df, purchase_amount, on="card_id", how='left')
test_df = pd.merge(test_df, purchase_amount, on='card_id', how='left')

month_lag = gdf['month_lag'].agg(['sum', 'mean', 'std', 'min', 'max']).reset_index()
month_lag.columns = ['card_id', 'hist_ml_sum', 'hist_ml_mean', 'hist_ml_std', 'hist_ml_min', 'hist_ml_max']

train_df = pd.merge(train_df, month_lag, on="card_id", how='left')
test_df = pd.merge(test_df, month_lag, on='card_id', how='left')

installments = gdf['installments'].agg(['sum', 'mean', 'std', 'min', 'max']).reset_index()
installments.columns = ['card_id', 'hist_in_sum', 'hist_in_mean', 'hist_in_std', 'hist_in_min', 'hist_in_max']

train_df = pd.merge(train_df, installments, on="card_id", how='left')
test_df = pd.merge(test_df, installments, on='card_id', how='left')
```

# Deal with missing values

```python
def check_missing_value(data):
    print(train_df.isna().sum() / train_df.shape[0])
```

```python
check_missing_value(train_df)
```

```python
def impute_missing_values(data):
    train_df['number_transactions'].fillna(0, inplace=True)
    
    col_fill_zero = ['hist_pa_sum', 'hist_pa_mean', 'hist_pa_std', 'hist_pa_min', 'hist_pa_max', 'hist_ml_sum', 'hist_ml_mean', 'hist_ml_std', 'hist_ml_min', 'hist_ml_max', 'hist_in_sum', 'hist_in_mean', 'hist_in_std', 'hist_in_min', 'hist_in_max']
    for col in col_fill_zero:
        data[col].fillna(0, inplace=True)
```

```python
impute_missing_values(train_df)
check_missing_value(train_df)

impute_missing_values(test_df)
check_missing_value(test_df)
```

# Transform features

```python
def transform_date(data):
    data['year'] = data['first_active_month'].dt.year
    data['month'] = data['first_active_month'].dt.month    
```

```python
transform_date(train_df)
transform_date(test_df)
```

```python
train_df.head(30)
```

```python
def select_features(data, cols):
    return data[cols]

#cols to be used for analysis
cols_to_use = ["feature_1", "feature_2", "feature_3", "year", "month", 
               "number_transactions",'hist_pa_sum', 'hist_pa_mean', 'hist_pa_std', 'hist_pa_min', 'hist_pa_max', 'hist_ml_sum', 'hist_ml_mean', 'hist_ml_std', 'hist_ml_min', 'hist_ml_max', 'hist_in_sum', 'hist_in_mean', 'hist_in_std', 'hist_in_min', 'hist_in_max']

train_X = select_features(train_df, cols_to_use)
test_X = select_features(test_df, cols_to_use)
train_y = train_df[target_col]
```

```python
test_X.head()
```

# Train the model

```python
estimator = lgb.LGBMRegressor(random_state=209,
                              silent=True,
                              subsample=0.7,
                              subsample_freq=50,
                              n_estimators=1000,
                              min_child_samples=500)
```

```python
N_FOLDS = 5
kf = KFold(n_splits=N_FOLDS, random_state=99, shuffle=True)
```

```python
#Define Param Grid
param_grid = {
     'learning_rate': [0.01, 0.1],
     'num_leaves':[10,20]
   }

grid = GridSearchCV(estimator, param_grid=param_grid, cv=kf, return_train_score=True)
```

```python
%%time

results = pd.DataFrame()
y_pred_gcv = 0

for dev_idx, val_idx in kf.split(train_df):
    train_X_GCV, val_X_GCV = train_X.loc[dev_idx, :], train_X.loc[val_idx, :]
    train_y_GCV, val_y_GCV = train_y.loc[dev_idx], train_y.loc[val_idx]
    
    grid.fit(train_X_GCV, train_y_GCV, eval_set=[(val_X_GCV, val_y_GCV)], eval_metric='rmse', early_stopping_rounds=100)
    results_tmp = pd.DataFrame(grid.cv_results_)
    results.append(results_tmp, ignore_index=True)
    
    y_pred_gcv_temp = grid.predict(test_X)
    y_pred_gcv += y_pred_gcv_temp

y_pred_gcv = y_pred_gcv / N_FOLDS * 1.0
```

```python
# print(results)
```

```python

submition = pd.DataFrame({'card_id': test_df['card_id'].values, 'target': y_pred_gcv})
submition.to_csv('submission', index=False)
```
