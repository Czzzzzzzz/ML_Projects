# Time series classification

# 1. Read Data

## Note
* Dataset comes from UCI AReM.

## Assignments
* read data
* create time-domain feature

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import itertools

import os

from pandas.plotting import scatter_matrix

from sklearn.utils import resample

from sklearn.preprocessing import label_binarize

from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve

from sklearn.linear_model import LogisticRegressionCV
from sklearn.multiclass import OneVsRestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
```

```python

def read_data(dir, prefix_fn, num, sep=','):
    full_path = os.path.join(dir, prefix_fn + str(num) + '.csv')    
#     print(full_path)
    dataset = pd.read_csv(full_path, header=None, comment='#', index_col=0, sep=sep)
#     print(dataset)    
    return dataset

def extract_feature(dataset):
        
    dataset_min = dataset.min(axis=0)
    dataset_max = dataset.max(axis=0)
    dataset_mean = dataset.mean(axis=0)
    dataset_median = dataset.median(axis=0)
    dataset_1st_quant = pd.DataFrame(dataset.quantile([0.25]).values.reshape(-1))
    dataset_3nd_quant = pd.DataFrame(dataset.quantile([0.75]).values.reshape(-1))
    
    dataset = pd.concat([dataset_min, dataset_max, dataset_mean, dataset_median, dataset_1st_quant, dataset_3nd_quant], ignore_index=True)

    return dataset

def process_data(dir, prefix_fn, num_range):
    dataset_list = []
    labels = []
    for dir_, n in zip(dir, num_range):
        for i in range(n[0], n[1]+1):
            labels.append(dir_)
            '''
            data stored in bending/dataset4 is separated by 
            whitespace instead of comma.
            '''
            if dir_ == 'AReM/bending2' and i == 4:
                temp_data = pd.DataFrame(read_data('AReM/bending2', prefix_fn, 4, ' ').values[:,:-1])
                dataset_list.append(extract_feature(temp_data))
                continue

            dataset_list.append(extract_feature(read_data(dir_, prefix_fn, i)))

    dataset = pd.concat(dataset_list, ignore_index=True, axis=1)
    dataset = np.transpose(dataset.values)
    
    return dataset, np.array(labels)

names = ['bending1', 'bending2', 'cycling', 'lying', 'sitting', 'standing', 'walking']
dir = ['AReM/' + name for name in names]
prefix = 'dataset'
traing_num = [[3, 7], [3, 6], [4, 15], [4, 15], [4,15], [4,15], [4,15]]
test_num = [[1, 2], [1, 2], [1, 3], [1, 3], [1, 3], [1, 3], [1, 3]]

'''
training set
'''
dataset, labels= process_data(dir, prefix, traing_num)
print(dataset.shape)
print(labels)

'''
testing set
'''
dataset_test, labels_test= process_data(dir, prefix, test_num)
print(dataset_test.shape)
```

# 2.Time-domain feature extraction

## Procedure
* compute the standard deviation of each of the domain features and estimate the confidence interval with the bootstrap
* select the three most important features.


```python
bootstrap_n = 10000
confidence_alpha = 2.5

sd = [np.std(resample(dataset), axis=0) for _ in range(bootstrap_n)]
print(np.array(sd))
```

```python
lower_bound = np.percentile(sd, confidence_alpha, axis=0)
# print(lower_bound.shape)
# print(lower_bound)

upper_bound = np.percentile(sd, 100 - confidence_alpha, axis=0)
# print(upper_bound)

std = np.std(dataset, axis=0)
sorted_index = np.argsort(std)
print(sorted_index)

sorted_std = np.array([std[i] for i in sorted_index])
print(sorted_std)

selected_feature_index = sorted_index[-12:]
```

```python
def transform_label(labels):
    result = labels.copy()
    for value, index in zip(labels, range(labels.shape[0])):
        if value == 'AReM/bending1' or value == 'AReM/bending2':
            result[index] = 1
        else:
            result[index] = 0
    return result

labels_2class = transform_label(labels)
labels_test_2class = transform_label(labels_test)

print(labels_2class)
print(labels_test_2class)
```

```python
%matplotlib inline

# df = pd.DataFrame(np.random.randn(100, 4), columns=['A', 'B', 'C', 'D'])
# df['target'] = np.repeat([1], 100)
# scatter_matrix(df, alpha=0.2, color=['r'], figsize=(16, 16))

def scatterplot_matrix_multiclass(data, labels, colors, names, **kwargs):
    """
    Sample input:
        fig = drawingTool.scatterplot_matrix_multiclass(data, label, ['r', 'b'], ['mpg', 'disp', 'drat', 'wt'], 
                marker='o')
    data: rows are features, colums are samples.
    
    Note:
        import matplotlib.pyplot as plt
        import itertools
        
    """
    
    numvars, numdata = data.shape
    fig, axes = plt.subplots(nrows=numvars, ncols=numvars, figsize=(16,16))
    fig.subplots_adjust(hspace=0.05, wspace=0.05)

    for ax in axes.flat:
        # Hide all ticks and labels
        ax.xaxis.set_visible(False)
        ax.yaxis.set_visible(False)

        # Set up ticks only on one side for the "edge" subplots...
        if ax.is_first_col():
            ax.yaxis.set_ticks_position('left')
        if ax.is_last_col():
            ax.yaxis.set_ticks_position('right')
        if ax.is_first_row():
            ax.xaxis.set_ticks_position('top')
        if ax.is_last_row():
            ax.xaxis.set_ticks_position('bottom')
    
    # Plot the data.
    for i, j in zip(*np.triu_indices_from(axes, k=1)):
        for x, y in [(i,j), (j,i)]:
            for i, (row_x_colval, row_y_colval) in enumerate(zip(data[x], data[y])):
                axes[x,y].scatter(row_x_colval, row_y_colval, color=colors[int(labels[i])], **kwargs)

    # Label the diagonal subplots...
    for i, label in enumerate(names):
        axes[i,i].annotate(label, (0.5, 0.5), xycoords='axes fraction',
                ha='center', va='center')

    # Turn on the proper x or y axes ticks.
    for i, j in zip(range(numvars), itertools.cycle((-1, 0))):
        axes[j,i].xaxis.set_visible(True)
        axes[i,j].yaxis.set_visible(True)

    return fig       

dataset_extracted = dataset[:, selected_feature_index]
print(dataset_extracted.shape)
# print(labels_2class)

fig = scatterplot_matrix_multiclass(dataset_extracted.T, labels_2class, ['r', 'b'], range(dataset_extracted.shape[1]), marker='o')
plt.show()
```

## break each time series into multiple equal length time series

```python
def extract_feature(dataset, n_series):
    feature_list = []
    interval = np.arange(0, dataset.values.shape[0]+1, int(dataset.values.shape[0] / n_series))
    for i in range(interval.shape[0]-1):
        sub_dataset = dataset.iloc[interval[i]:interval[i+1], :]
        dataset_min = sub_dataset.min(axis=0)
        dataset_max = sub_dataset.max(axis=0)
        dataset_mean = sub_dataset.mean(axis=0)
#         dataset_median = sub_dataset.median(axis=0)
#         dataset_1st_quant = pd.DataFrame(sub_dataset.quantile([0.25]).values.reshape(-1))
#         dataset_3nd_quant = pd.DataFrame(sub_dataset.quantile([0.75]).values.reshape(-1))        
        feature_list.append(dataset_min)
        feature_list.append(dataset_max)
        feature_list.append(dataset_mean)
    
    dataset = pd.concat(feature_list, ignore_index=True)
    return dataset

def process_data(dir, prefix_fn, num_range, n_series):
    dataset_list = []
    labels = []
    for dir_, n in zip(dir, num_range):
        for i in range(n[0], n[1]+1):
            labels.append(dir_)
            '''
            data stored in bending/dataset4 is separated by 
            whitespace instead of comma.
            '''
            if dir_ == 'AReM/bending2' and i == 4:
                temp_data = pd.DataFrame(read_data('AReM/bending2', prefix_fn, 4, ' ').values[:,:-1])
                dataset_list.append(extract_feature(temp_data, n_series))
                continue

            dataset_list.append(extract_feature(read_data(dir_, prefix_fn, i), n_series))

    dataset = pd.concat(dataset_list, ignore_index=True, axis=1)
    dataset = np.transpose(dataset.values)
    
    return dataset, np.array(labels)

names = ['bending1', 'bending2', 'cycling', 'lying', 'sitting', 'standing', 'walking']
dir = ['AReM/' + name for name in names]
prefix = 'dataset'
traing_num = [[3, 7], [3, 6], [4, 15], [4, 15], [4,15], [4,15], [4,15]]
test_num = [[1, 2], [1, 2], [1, 3], [1, 3], [1, 3], [1, 3], [1, 3]]

# '''
# training set
# '''
# dataset, labels= process_data(dir, prefix, traing_num, 2)
# print(dataset.shape)
# # print(labels)

# '''
# testing set
# '''
# dataset_test, labels_test= process_data(dir, prefix, test_num, 2)
# print(dataset_test.shape)
```

# 3. model selection and train

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression

from sklearn.feature_selection import RFE
from sklearn.feature_selection import RFECV


avg_scores = np.array([])
    
l_splits_num = np.arange(1, 21)
for i in l_splits_num:
    dataset, labels= process_data(dir, prefix, traing_num, i)
    print(dataset.shape)

    dataset_test, labels_test= process_data(dir, prefix, test_num, i)
    print(dataset_test.shape)

    avg_score = 0
    n_folds = 5
    skf = StratifiedKFold(n_splits=n_folds)
    for train_index, test_index in skf.split(dataset, labels_2class):
        X_train, X_test = dataset[train_index], dataset[test_index]
        y_train, y_test = labels_2class[train_index], labels_2class[test_index]

        rfecv = RFECV(estimator=LogisticRegression(), step=1, cv=StratifiedKFold(2), scoring='accuracy')
        rfecv.fit(X_train, y_train)

    #     print("Optimal number of features: %d" % rfecv.n_features_)            
        X_train_selection = rfecv.transform(X_train)
        X_test_selection = rfecv.transform(X_test)

        lr = LogisticRegression()
        lr.fit(X_train_selection, y_train)
    #     lr.predict(X_test)
        avg_score += lr.score(X_test_selection, y_test)
    avg_scores = np.append(avg_scores, avg_score / n_folds)
#     avg_scores.append(avg_score / n_folds)
    
    print(i)
    
plt.plot(l_splits_num, avg_scores)
plt.show()
```

```python
optimal_splits_num = np.argmax(avg_scores) + 1

dataset, labels= process_data(dir, prefix, traing_num, 1)
dataset_test, labels_test= process_data(dir, prefix, test_num, 1)

labels_2class = transform_label(labels)
labels_test_2class = transform_label(labels_test)

lr = LogisticRegression()
rfecv = RFECV(estimator=lr, step=1, cv=StratifiedKFold(2), scoring='accuracy')
rfecv.fit(dataset, labels_2class)
print(rfecv.n_features_)

dataset_selection = rfecv.transform(dataset)
dataset_test_selection = rfecv.transform(dataset_test)

lr.fit(dataset_selection, labels_2class)

print('coefficients')
print(lr.coef_)

score = lr.score(dataset_test_selection, labels_test_2class)
print('accuracy: %f' % score)

labels_predicted = lr.predict(dataset_test_selection)
tn, fp, fn, tp = confusion_matrix(labels_test_2class, labels_predicted).ravel()
print('confusion matrix: (tn, fp, fn, tp)')
print((tn, fp, fn, tp))

labels_predicted_score = lr.decision_function(dataset_test_selection)

fpr_rt_lm, tpr_rt_lm, _ = roc_curve(labels_test_2class, labels_predicted_score, pos_label='1')

plt.figure()
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr_rt_lm, tpr_rt_lm)
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.title('ROC')
plt.show()
```

## 4.1 multinomial regression model

```python
avg_scores = np.array([])
    
l_splits_num = np.arange(1, 21)
for i in l_splits_num:
    dataset, labels= process_data(dir, prefix, traing_num, i)
#     print(dataset.shape)

#     dataset_test, labels_test= process_data(dir, prefix, test_num, i)
#     print(dataset_test.shape)
    
    avg_score = 0
    n_folds = 5
    skf = StratifiedKFold(n_splits=n_folds)
    for train_index, test_index in skf.split(dataset, labels):
#         print(test_index)
        X_train, X_test = dataset[train_index], dataset[test_index]
        y_train, y_test = labels[train_index], labels[test_index]

        lr = LogisticRegressionCV(penalty='l1', solver='saga', multi_class='multinomial')
        lr.fit(X_train, y_train)
        avg_score += lr.score(X_test, y_test)
    avg_scores = np.append(avg_scores, avg_score / n_folds)
    
    print(i)
    
plt.plot(l_splits_num, avg_scores)
plt.show()
```

## 4.2 Gaussian Naive Bayes and Multinomial Native Bayes

```python
classifiers = [GaussianNB(), MultinomialNB()]
optimal_splits_nums = []

for clf in classifiers:

    avg_scores = np.array([])
    l_splits_num = np.arange(1, 21)
    for i in l_splits_num:
        dataset, labels= process_data(dir, prefix, traing_num, i)

        avg_score = 0
        n_folds = 5
        skf = StratifiedKFold(n_splits=n_folds)
        for train_index, test_index in skf.split(dataset, labels):
    #         print(test_index)
            X_train, X_test = dataset[train_index], dataset[test_index]
            y_train, y_test = labels[train_index], labels[test_index]

            lr = clf
            lr.fit(X_train, y_train)
            avg_score += lr.score(X_test, y_test)
        avg_scores = np.append(avg_scores, avg_score / n_folds)

        print(i)

    plt.plot(l_splits_num, avg_scores)
    plt.show()
    
    optimal_splits_num = np.argmax(avg_scores) + 1
    optimal_splits_nums.append(optimal_splits_num)
    
    
```
